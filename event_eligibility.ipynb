```python
import pandas as pd

df = pd.read_excel("/content/events_dataset_cleaned_names_links (1).xlsx")
```

```python
df.columns = df.columns.str.strip()
df["skills"] = df["skills"].str.lower()
df["region"] = df["region"].str.lower()
df["mode"] = df["mode"].str.lower()
df["type"] = df["type"].str.lower()
```

```python
def check_eligibility(student_gpa, student_region, student_skills, student_mode=None):

    student_region = student_region.lower()
    student_skills = [s.lower().strip() for s in student_skills]

    eligible = []

    for index, row in df.iterrows():

        # GPA rule
        if student_gpa < row["min_gpa"]:
            continue

        # Region rule
        if not (row["region"] == "national" or row["region"] == student_region):
            continue

        # Skills rule
        event_skills = [s.strip() for s in row["skills"].split(",")]
        if not any(skill in event_skills for skill in student_skills):
            continue

        # Mode rule (optional)
        if student_mode:
            if row["mode"] != student_mode.lower():
                continue

        eligible.append(row["name"])

    return eligible
```

```python
results = check_eligibility(
    student_gpa=7.5,
    student_region="kerala",
    student_skills=["science", "coding"],
    student_mode="offline"
)

import pandas as pd

results_df = pd.DataFrame(results)
results_df
```
